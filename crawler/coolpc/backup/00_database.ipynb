### 使用流程
1. 建立資料庫、表格

```python
# # # 建立資料庫
import sqlite3  
import pandas as pd
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
  
# print("Opened database successfully")  
```

```python
# 建立表格
table_name = 'mainboard'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID            INT , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL          TEXT    ,
       BRAND          TEXT    ,
       ETL_DATE           DATETIME
       );''')


table_name = 'ram'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID             INT , 
       NAME           TEXT    , 
       PRICE          INT   ,
       MODEL          TEXT   ,
       BRAND          TEXT   ,
        ETL_DATE           DATETIME
       );''')
  

table_name = 'hdd'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID            INT      , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL          TEXT    ,
       BRAND          TEXT    ,
        ETL_DATE           DATETIME
       );''')


table_name = 'cpu'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID             INT     , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL          TEXT    ,
       BRAND          TEXT    ,
       ETL_DATE           DATETIME  
       );''')


table_name = 'gpu'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID INT , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL          TEXT    ,
       BRAND          TEXT    ,
       ETL_DATE       DATETIME
       );''')
conn.close()  
```

```python
# 建立散熱器的TABLE
table_name='radiator'
conn.execute(f'''CREATE TABLE IF NOT EXISTS {table_name} 
       (ID INT , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL        TEXT,
       BRAND          TEXT    ,
       ETL_DATE       DATETIME,
       TDP          TEXT    ,
       RPM          TEXT    ,
       TYPE         TEXT
       );''')
conn.close()  
```

```python
#丟掉表格
import sqlite3  
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
  
table_name='gpu'
conn.execute(f'''DROP TABLE {table_name};
           ''')
conn.close()  
```

## 增加欄位/減少欄位

```python
# 加入特定欄位
import sqlite3  
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
  
table_name='ram'
conn.execute(f'''ALTER TABLE {table_name} 
           ADD  COLUMN 
           capacity_unit varchar;
           ''')
conn.close()  
```

```python
# 丟掉特定欄位
# # # 建立資料庫
import sqlite3  
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
  
table_name='CPU'
conn.execute(f'''ALTER TABLE {table_name} 
           drop  COLUMN 
           performance;
           ''')
conn.close()  
```

```python
#增加表格
import sqlite3  
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
table_name = 'motherboard'
conn.execute(f'''CREATE TABLE {table_name} 
       (ID INT , 
       NAME           TEXT    , 
       PRICE          INT    ,
       MODEL          TEXT    ,
       BRAND          TEXT    ,
       ETL_DATE           DATETIME
       );''')
conn.close()  
```
