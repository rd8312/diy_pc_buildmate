### 1. 建立Creat、Insert、Drop、Add等各項資料庫操作function

```python
# # # 建立資料庫
import sqlite3  
import pandas as pd
  
db_name = 'data/sqlchain.db'
conn = sqlite3.connect(db_name)  
  
# print("Opened database successfully")  
```

```python
def createTableWithColumns(db_name, table_name, columns):
    """
    
    如果指定的表不存在，則在SQLite資料庫中根據提供的列定義創建該表。
    參數:
    db_name (str): 資料庫檔案名。
    table_name (str): 要創建的表名。
    columns (str): 表的列定義，列名和資料類型應按SQL語法給出，列與列之間用逗號分隔。
    """
    try:
        conn = sqlite3.connect(db_name)
        create_table_sql = f'''CREATE TABLE IF NOT EXISTS {table_name} ({columns});'''
        conn.execute(create_table_sql)
        conn.commit()
    except sqlite3.Error as error:
        print(f"数据库操作出错: {error}")
    finally:
        conn.close()
```

```python
import sqlite3

def dropTable(db_name, table_name):
    """
    删除指定的SQLite資料表。

    参数:
    db_name (str): 資料表文件名。
    table_name (str): 要删除的表名。
    """
    try:
        conn = sqlite3.connect(db_name)
        cursor = conn.cursor()
        cursor.execute(f"DROP TABLE IF EXISTS {table_name}")
        conn.commit()
    except sqlite3.Error as error:
        print(f"資料表操作出错: {error}")
    finally:
        conn.close()

# 使用示例
#dropTable(db_name, 'motherboard')
```

```python
def addColumn(db_name, table_name, column_definition):
    """
    向指定的SQLite表中添加一个新列。

    参数:
    db_name (str): 資料表名。
    table_name (str): 要修改的表名。
    column_definition (str): 新列的定義，包括列名和數據類型，如`"new_column TEXT"`。
    """
    try:
        conn = sqlite3.connect(db_name)
        cursor = conn.cursor()
        cursor.execute(f"ALTER TABLE {table_name} ADD COLUMN {column_definition}")
        conn.commit()
    except sqlite3.Error as error:
        print(f"資料表操作出錯: {error}")
    finally:
        conn.close()

# 使用示例
#addColumn(db_name, 'battery', 'tdp INT')
```

```python
def dropColumn(db_name, table_name, column_definition):
    """
    在SQLite資料表中丟掉欄位

    参数:
    db_name (str): 資料庫文件名。
    table_name (str): 要修改的表名。
    column_definition (str): 新列的定義，包括列名，如`"new_column"`。
    """
    try:
        conn = sqlite3.connect(db_name)
        cursor = conn.cursor()
        cursor.execute(f"ALTER TABLE {table_name} drop  COLUMN {column_definition}")
        conn.commit()
    except sqlite3.Error as error:
        print(f"資料表操作出錯: {error}")
    finally:
        conn.close()

dropColumn(db_name, 'mainboard', 'model')
```
